# Visualizing the Distribution of Population 

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('population.csv')
```

```python
year = '2023'

# Ensure the column exists and remove rows with missing values
if year in df.columns:
    df[year] = pd.to_numeric(df[year], errors='coerce')  # Convert to numeric, handling errors
    population_data = df[year].dropna()  # Drop missing values

    # Create a histogram for population distribution
    plt.figure(figsize=(10, 6))
    sns.histplot(population_data, bins=30, kde=True, color='blue')
    plt.title(f'Population Distribution in {year}')
    plt.xlabel('Population')
    plt.ylabel('Frequency')
    plt.show()
else:
    print(f"Column '{year}' not found in the dataset.")
```

###### The population distribution in 2023 reveals a strikingly uneven global landscape, with a clear divide between the handful of populous nations and the majority of countries. This histogram shows that most countries cluster at the lower end of the population scale, while just a few outliers stretch into the billions, creating a dramatic right-skewed distribution. It's a powerful visualization of how concentrated the world's population is in a small number of nations, while most countries maintain relatively modest population sizes under 500 million.

```python

year = '2023'

# clean data
if year in df.columns:
    df[year] = pd.to_numeric(df[year], errors='coerce')
    top_countries = df[['Country Name', year]].dropna().sort_values(by=year, ascending=False).head(10)

    # Bar chart for top 10 countries
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_countries[year], y=top_countries['Country Name'], palette='viridis')
    plt.title(f'Top 10 Countries by Population in {year}')
    plt.xlabel('Population')
    plt.ylabel('Country')
    plt.show()
else:
    print(f"Column '{year}' not found in the dataset.")
```

###### This bar chart offers a fascinating breakdown of global population segments in 2023, highlighting key economic and demographic divisions. The world's total population tops around 8 billion, with a significant portion falling into low & middle income categories. The distinction between "early-demographic dividend" and "late-demographic dividend" countries reveals where different regions are in their population transition phases. The chart also shows how international development institutions (IDA & IBRD) categorize large population segments, with the East Asia & Pacific region notably representing a substantial portion of the global population. The gradient in income levels, from upper middle to lower middle income, illustrates the economic diversity within the global population distribution.

```python
# Select multiple countries
countries = ['India', 'China', 'United States', 'Indonesia', 'Pakistan']
comparison_data = df[df['Country Name'].isin(countries)][['Country Name', '2020']]

# Bar chart for population comparison
plt.figure(figsize=(10, 6))
sns.barplot(x='Country Name', y='2020', data=comparison_data, palette='coolwarm')
plt.title('Population Comparison in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.show()
```

###### This 2020 population comparison reveals the stark population differences between major nations. China and India stand out dramatically, each with approximately 1.4 billion people, dwarfing the other countries shown. Indonesia, with about 270 million people, represents a significant but much smaller population, while Pakistan and the United States follow with populations around 220 million and 330 million respectively. The graph effectively illustrates the massive demographic weight of Asia's two largest nations, which together account for a substantial portion of the world's population, while showing how even large countries like the United States have far smaller populations in comparison.

```python

start_year = '2000'
end_year = '2023'

#calculate growth rate
if start_year in df.columns and end_year in df.columns:
    df[start_year] = pd.to_numeric(df[start_year], errors='coerce')
    df[end_year] = pd.to_numeric(df[end_year], errors='coerce')
    df['Growth Rate (%)'] = ((df[end_year] - df[start_year]) / df[start_year]) * 100

    # Top 10 countries with the highest growth rate
    top_growth = df[['Country Name', 'Growth Rate (%)']].dropna().sort_values(by='Growth Rate (%)', ascending=False).head(10)

    # Bar chart for growth rate
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Growth Rate (%)', y='Country Name', data=top_growth, palette='crest')
    plt.title(f'Top 10 Countries by Population Growth Rate ({start_year} - {end_year})')
    plt.xlabel('Growth Rate (%)')
    plt.ylabel('Country')
    plt.show()
else:
    print(f"One or both columns '{start_year}' and '{end_year}' not found in the dataset.")
```

###### This chart reveals fascinating population growth patterns from 2000 to 2023, with a striking dominance of Middle Eastern nations and select African countries. Qatar leads with an exceptional growth rate of around 300%, followed by the UAE at about 250%. What's particularly interesting is the clear divide between two groups: Gulf states (Qatar, UAE, Kuwait, Bahrain, and Oman) driven by economic migration and development, and African nations (Equatorial Guinea, Niger, Chad, Angola) experiencing high natural population growth. The Turks and Caicos Islands stands out as the only non-Middle Eastern or African territory in this top 10, likely due to its development as a tourism and financial hub. This pattern highlights how both economic opportunity and natural demographic trends can drive rapid population expansion.

```python
# Extract year columns (1960-2023)
year = df.columns[4:]  
total_population= df[year].sum().dropna()

# Plotting the global population trend over the years using bar chart
plt.figure(figsize=(14, 7))
plt.bar(total_population.index, total_population.values,color='black')
plt.title('Population Trends (1960-2023)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Population', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

###### This graph depicts global population growth from 1960 to 2023, showing a remarkable upward trend from about 3 billion to over 8 billion people. The trend line reveals consistent growth but with a subtle yet important change in its slope - while growth has been continuous, there's evidence of a gradual slowing in the rate of increase in recent years. What's particularly noteworthy is that despite this deceleration, the absolute population continues to climb significantly. The steady, almost linear ascent speaks to the profound demographic transformation that has occurred over these six decades, reflecting improvements in healthcare, longer life expectancy, and changing family patterns worldwide.
